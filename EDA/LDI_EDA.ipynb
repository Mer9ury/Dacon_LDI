```
ted = pd.read_csv('./dataset/train_err_data.csv',thousands = ',')
tqd = pd.read_csv('./dataset/train_quality_data.csv',thousands = ',')
tpd = pd.read_csv('./dataset/train_problem_data.csv',thousands = ',')
```

```
with open('./Preprocessed_data/v3/traindata.txt', 'rb') as f:
    X = pickle.load(f)
```

```
df_tpd = tpd.drop_duplicates()
problem_list = list(df_tpd['user_id'].unique())
y = np.zeros(15000,)
for pi in problem_list:
    y[pi-10000] = 1
```

```
columns = []
for i in range(1,43):
    s = 'err_' + str(i)
    columns.append(s)
columns.append('23_ct')
columns.append('23_as')
columns.append('23_ac')
columns.append('31_0')
columns.append('40_0')
for i in range(13):
    s = 'quality_' + str(i)
    columns.append(s)
columns.append('result')
a = np.insert(X,60,y,axis=1)
df = pd.DataFrame(a,columns=columns)
df.head()
```
